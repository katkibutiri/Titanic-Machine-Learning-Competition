```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
```

# Loading Housing Data



```python
data = pd.read_csv("train.csv")
```

```python
data.head(10)
```

```python
from sklearn.impute import SimpleImputer

# Assuming "Age" is the column you want to impute
age_imputer = SimpleImputer(strategy='mean')

# Fit the imputer on the entire dataset
age_imputer.fit(data[['Age']])

# Impute missing values in the dataset
data['Age'] = age_imputer.transform(data[['Age']])

# Round up the imputed age values
data['Age'] = data['Age'].apply(lambda x: round(x))
```

```python
columns_to_drop = ['Name', 'SibSp', 'Ticket', 'Cabin', 'Embarked','PassengerId']
data.drop(columns=columns_to_drop, inplace=True)
data
```

```python
data.info()
```

```python
data = pd.get_dummies(data, columns=['Sex'])

data
```

```python
# Assuming data_dropped_vif is your feature matrix and target variable is 'Survived'
X_train = data.drop(columns=["Survived"])
y_train = data["Survived"]

from sklearn.ensemble import RandomForestClassifier

# Create the Random Forest Classifier
rf_classifier = RandomForestClassifier()

# Train the model
rf_classifier.fit(X_train, y_train)
```

```python
# Load the test data
test_data = pd.read_csv('test.csv')
test_data
```

```python
test_data.info()
```

```python
from sklearn.impute import SimpleImputer

# Assuming "Age" is the column you want to impute
age_imputer = SimpleImputer(strategy='mean')

# Fit the imputer on the entire dataset
age_imputer.fit(test_data[['Age']])

# Impute missing values in the dataset
test_data['Age'] = age_imputer.transform(test_data[['Age']])

# Round up the imputed age values
test_data['Age'] = test_data['Age'].apply(lambda x: round(x))


# Assuming "Age" is the column you want to impute
age_imputer = SimpleImputer(strategy='mean')

# Fit the imputer on the entire dataset
age_imputer.fit(test_data[['Fare']])

# Impute missing values in the dataset
test_data['Fare'] = age_imputer.transform(test_data[['Fare']])

# Round up the imputed age values
test_data['Fare'] = test_data['Fare'].apply(lambda x: round(x))
```

```python
columns_to_drop = ['Name', 'SibSp', 'Ticket', 'Cabin', 'Embarked','PassengerId']
test_data.drop(columns=columns_to_drop, inplace=True)
test_data
```

```python
test_data = pd.get_dummies(test_data, columns=['Sex'])
test_data.info()
```

```python
# Predict on the test set
y_pred_rf = rf_classifier.predict(test_data)

# Get the IDs from the test data
test = pd.read_csv('test.csv')
test_ids = test['PassengerId']

# Create a DataFrame with IDs and predicted prices
predictions_df = pd.DataFrame({'PassengerId': test_ids, 'Survived': y_pred_rf})

# Save the predictions to a CSV file
predictions_df.to_csv('Submission.csv', index=False)
```
